```python
import subprocess

image_path = 'C:\\Users\\salmo\\Downloads\\audioToVideo\\room1.jpg'
audio_path = 'C:\\Users\\salmo\\Downloads\\audioToVideo\\audio.mp3'
output_path = 'C:\\Users\\salmo\\Downloads\\audioToVideo\\video.mp4'

ffmpeg_path = r'C:\Users\salmo\Downloads\audioToVideo\ffmpeg\ffmpeg\bin\ffmpeg.exe'

command = [
    ffmpeg_path,
    '-loop', '1',
    '-i', image_path,
    '-i', audio_path,
    '-c:v', 'libx264',
    '-pix_fmt', 'yuv420p',
    '-c:a', 'aac',
    '-b:a', '192k',
    '-shortest',
    output_path
]

# Run the command and capture any errors
result = subprocess.run(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

# Print FFmpeg output (if any)
print(result.stdout)
print(result.stderr)
```
